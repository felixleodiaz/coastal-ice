```python
# read in libraries

import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import random
from pathlib import Path
from tqdm import tqdm
```

```python
# read in csv

folderpath = Path('../local_data/SeaIce_Results_ScatterData')
files = list(folderpath.glob('*.csv'))
save_path = Path('../local_data/SeaIce_Results_ScatterPlots')

# create nir-green-blue vectors (magnitude, direction)

for file in tqdm(files, desc="Generating scatter plots"):
    df = pd.read_csv(file)
    plt.figure(figsize=(6, 6))

    sns.scatterplot(data=df, x='blue', y='nir', hue='cluster_label', 
                    legend='auto', alpha = 0.6)
    plt.xlabel('Blue Reflectance')
    plt.ylabel('NIR Reflectance')
    plt.title(file.stem)

    plt.legend(title='Cluster', markerscale=2)
    
    plt.savefig(save_path / f'{file.stem}.png', dpi=300, bbox_inches='tight')
    plt.close()
```

```python
# create 3D scatter

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')

ax.scatter(file['blue'], file['green'], file['nir'])
plt.show()
```

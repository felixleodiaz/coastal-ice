```python
from osgeo import gdal
from pathlib import Path
from tqdm import tqdm
```

```python
# cloud images

input_dir  = Path("../../local_data/CloudMaskExamples")
output_dir = Path("../../local_data/CloudMaskExamples_PNG")
output_dir.mkdir(exist_ok=True)

for tif in tqdm(list(input_dir.glob("*.tif")), desc="Converting"):
    gdal.Translate(
        str(output_dir / f"{tif.stem}.png"),
        str(tif),
        format="PNG"
    )
```
